```python
import numpy as np
import matplotlib.pyplot as plt
from pytplot import tplot_restore, get_data
from datetime import datetime

# Define the path and filename for November 2nd
path2 = '/Users/vanishaswabhanam/Downloads/tplots/'
filename2 = 'CIRBE_data_20231102.tplot'

# Restore the tplot data for November 2nd
tplot_restore(path2 + filename2)

# Events from November 2nd
events_nov2 = [
    {"start": "2023-11-02 01:25:00", "end": "2023-11-02 01:44:00", "IB": -68, "MLT": 23.7},
    {"start": "2023-11-02 02:57:00", "end": "2023-11-02 03:20:00", "IB": -67, "MLT": 23.8},
    {"start": "2023-11-02 04:07:00", "end": "2023-11-02 04:20:00", "IB": 66, "MLT": 23.9},
    {"start": "2023-11-02 06:16:00", "end": "2023-11-02 06:24:00", "IB": -67, "MLT": 22.2},
    {"start": "2023-11-02 07:50:00", "end": "2023-11-02 07:58:00", "IB": -67, "MLT": 22.0},
    {"start": "2023-11-02 08:52:00", "end": "2023-11-02 09:01:00", "IB": 67, "MLT": 23.4}
]

# Function to convert a time string to a UNIX timestamp
def to_timestamp(time_str):
    return datetime.strptime(time_str, "%Y-%m-%d %H:%M:%S").timestamp()

# Function to calculate cell edges for pcolormesh
def calculate_edges(values):
    edges = np.zeros(len(values) + 1)
    edges[1:-1] = (values[:-1] + values[1:]) / 2.0
    edges[0] = values[0] - (values[1] - values[0]) / 2.0
    edges[-1] = values[-1] + (values[-1] - values[-2]) / 2.0
    return edges

# Create the figure and subplots with manual layout adjustment
fig, axes = plt.subplots(len(events_nov2), 1, figsize=(8, 12), gridspec_kw={'hspace': 0.6})
im = None

# Loop through each event to create subplots
for i, event in enumerate(events_nov2):
    start_time = to_timestamp(event['start'])
    end_time = to_timestamp(event['end'])
    IB = event['IB']
    MLT = event['MLT']
    
    # Get the data
    time_flux, flux, energy = get_data('cirbe_efluxe_adapted')
    time_MLATaacgm, MLAT_aacgm = get_data('cirbe_MLAT_aacgm')
    
    # Convert the time to datetime for filtering
    standard_time = [datetime.utcfromtimestamp(t) for t in time_flux]
    
    # Filter the data based on the time range
    filtered_indices = [idx for idx, t in enumerate(standard_time) if start_time <= t.timestamp() <= end_time]
    
    if len(filtered_indices) == 0:
        print(f"No data found for event {i+1}: Check time range or data availability")
        continue
    
    # Get the filtered data
    filtered_flux = flux[filtered_indices]
    filtered_energy = energy
    MLAT_filtered = MLAT_aacgm[filtered_indices]
    
    # Ensure no non-finite values exist (NaNs or Infs) in the data
    mask_flux = np.isfinite(filtered_flux).all(axis=1)
    mask_MLAT = np.isfinite(MLAT_filtered)
    
    # Apply the mask to ensure finite values only
    valid_mask = mask_flux & mask_MLAT
    
    if np.sum(valid_mask) == 0:
        print(f"No valid data found for event {i+1} after removing non-finite values.")
        continue
    
    # Use only valid data points
    filtered_flux = filtered_flux[valid_mask]
    MLAT_filtered = MLAT_filtered[valid_mask]
    
    # Handle potential negative or zero values in flux/energy
    filtered_flux[filtered_flux <= 0] = np.nan
    filtered_energy[filtered_energy <= 0] = np.nan

    # Remove NaN values before plotting
    valid_energy_mask = np.isfinite(filtered_energy)
    filtered_energy = filtered_energy[valid_energy_mask]
    filtered_flux = filtered_flux[:, valid_energy_mask]
    
    # Calculate edges for MLAT and energy
    MLAT_edges = calculate_edges(MLAT_filtered)
    energy_edges = calculate_edges(filtered_energy)

    if len(MLAT_filtered) > 0 and len(filtered_flux) > 0:
        # Create a spectrogram-like plot
        im = axes[i].pcolormesh(MLAT_edges, np.log10(energy_edges), np.log10(filtered_flux).T, shading='auto', cmap='jet')
        
        # Plot the IB boundary as a vertical line
        axes[i].axvline(x=IB, color='black', linestyle='--', label=f'IB={IB}')
        
        # Add MLT annotation
        axes[i].text(0.05, 0.95, f"<MLT>={MLT}", transform=axes[i].transAxes, fontsize=12, verticalalignment='top')
        
        axes[i].set_title(f'CIRBE {event["start"][:10]} {event["start"][11:16]}-{event["end"][11:16]} <MLT>={MLT}', fontsize=12)
        
        # Labeling and formatting
        axes[i].set_ylabel('Energy [MeV]')
        axes[i].set_xlim([min(MLAT_filtered)-5, max(MLAT_filtered)+5])
        axes[i].set_ylim([np.log10(min(filtered_energy)), np.log10(max(filtered_energy))])
        axes[i].legend()
    
    if i == len(events_nov2) - 1:
        axes[i].set_xlabel('MLAT')

# Move the colorbar to the side of the entire figure
if im is not None:
    fig.subplots_adjust(right=0.85)
    cbar_ax = fig.add_axes([0.87, 0.15, 0.03, 0.7])
    fig.colorbar(im, cax=cbar_ax, label='Flux [cm^2/s/sr/MeV]')

plt.show()
```
